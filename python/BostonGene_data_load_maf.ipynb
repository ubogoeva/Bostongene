```python
import requests
import pandas as pd
import io
import os

#upload data to cloud
url = 'https://www.cbioportal.org/webservice.do'
all_genes_df = pd.read_csv('../data/Mutated_Genes_laml.tsv', sep = '\t')
genes_filtered = all_genes_df[all_genes_df['# Mut'] > 1].iloc[:,0]
genes_filtered
```

```python
#define function to parse data from portal to pandas dataframe
def parse_data(req_result):
    f = io.StringIO(req_result.text)
    df = []
    for line in f.readlines():
        line = line.replace('\n', '')
        if line.startswith('#'):
            continue               
        elif line.startswith('entrez'):    
            columns = line.split('\t')
        else:
            df.append(line.split('\t'))
    parsed_data = pd.DataFrame(df, columns = columns)
    return parsed_data
```

```python
laml_maf = pd.DataFrame()
print('enter mutation study id')
mut_id = str(input())
for gene in genes_filtered:
    payload = {'cmd': 'getMutationData', 'genetic_profile_id':mut_id, 'gene_list': gene}
    r = requests.get(url, params=payload)
    print(r.status_code ,gene, end='\t')
    temp_df = parse_data(r)
    laml_maf = laml_maf.append(temp_df)
laml_maf.to_csv('../data/genes_sign_mutsig_laml.csv', index = False)
print('\n', 'data loaded successfully')
```
